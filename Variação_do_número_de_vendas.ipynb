```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
vendas_2022 = pd.DataFrame(vendas_2022, index=lojas)
```

```python
display(vendas_2022)
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.suptitle("Vendas em todo o período de 2022")
plt.subplots_adjust(wspace=0.3, hspace=0.4)

axs[0,0].plot(vendas_2022.loc['A'])
axs[0,0].set_title("Vendas da empresa A")

axs[0,1].plot(vendas_2022.loc['B'])
axs[0,1].set_title("Vendas da empresa B")

axs[1,0].plot(vendas_2022.loc['C'])
axs[1,0].set_title('Vendas da empresa C')

axs[1,1].plot(vendas_2022.loc['D'])
axs[1,1].set_title('Vendas da empresa D')

for ax in axs.flat:
  ax.set_xlabel("Mês")
  ax.set_ylabel("Número de vendas")

plt.show()
```
